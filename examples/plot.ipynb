# suneku.plot

```python
from pandas import read_csv
from suneku import REPO, hello, plot
```

```python
DATADIR = REPO/'examples/data'
ENERGY = DATADIR/'NewYorkEnergy.csv'
TEMPS = DATADIR/'ZonalTempAnomaly.csv'
```

```python
hello(plot)
```

## read data from CSV files

```python
def read(path,index):
    print("Read {}".format(path))
    data = read_csv(path,index_col=index,dtype=float)    
    return data.sort_index().sort_index(axis=1)
```

**energy:**
GWh of electricity generated in New York since 1980 from
[data.ny.gov](https://data.ny.gov/Energy-Environment/Electric-Generation-By-Fuel-Type-GWh-Beginning-196/h4gs-8qnu)

```python
energy = read(ENERGY,'Year').drop(columns=['Total'])
energy.tail()
```

**temps:**
Global surface temperature anomalies since 1880 from
[data.giss.nasa.gov](https://data.giss.nasa.gov/gistemp/)

```python
temps = read(TEMPS,'Year').drop(columns=['Glob','NHem','SHem'])
temps.tail()
```

## area

```python
hello(plot.area)
plot.area(energy)
```

## bar

```python
hello(plot.bar)
plot.bar(energy)
```

## hist

```python
hello(plot.hist)
plot.hist(temps)
```

## line

```python
hello(plot.line)
plot.line(temps)
```

## quantile

```python
hello(plot.quantile)
plot.quantile(temps)
```
