# suneku.plot

```python
from pandas import read_csv
from suneku import REPO, hello, plot
hello(plot)
```

```python
DATADIR = REPO/'examples/data'
ENERGY = DATADIR/'NewYorkEnergy.csv'
TEMPS = DATADIR/'ZonalTempAnomaly.csv'
```

## read data from CSV files

**energy:**
GWh of electricity generated in New York since 1980 from
[data.ny.gov](https://data.ny.gov/Energy-Environment/Electric-Generation-By-Fuel-Type-GWh-Beginning-196/h4gs-8qnu)

```python
energy = ( read_csv(ENERGY,index_col='Year')
        .drop(columns=['Total'])
        .astype(float) )

energy.tail()
```

**temps:** Global surface temperature anomalies since 1880 from
[data.giss.nasa.gov](https://data.giss.nasa.gov/gistemp/)

```python
temps = ( read_csv(TEMPS,index_col='Year')
        .drop(columns=['Glob','NHem','SHem'])
        .astype(float) )
    
temps.tail()
```

## area

```python
hello(plot.area)
fig = plot.area(energy)
```

## bar

```python
hello(plot.bar)
fig = plot.bar(energy)
```

## hist

```python
hello(plot.hist)
fig = plot.hist(temps)
```

## line

```python
hello(plot.line)
fig = plot.line(temps)
```

## quantile

```python
hello(plot.quantile)
fig = plot.quantile(temps)
```
